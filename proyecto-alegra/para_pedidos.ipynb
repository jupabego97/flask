```python
import pandas as pd

facturas = pd.read_csv('facturas.csv')

items = pd.read_csv('items.csv',sep=',', encoding='utf-8')

proveedor = pd.read_csv('facturas_proveedor.csv')
```

```python
import datetime

current_date = datetime.date.today()

# Calculate the date of yesterday


anterior1 = current_date - datetime.timedelta(days=90)

anterior2 = current_date - datetime.timedelta(days=30)

anterior3 = current_date - datetime.timedelta(days=7)




```

```python



facturas_90 = facturas[facturas['fecha'] > str(anterior1)]

facturas_30 = facturas[facturas['fecha'] > str(anterior2)]

facturas_7= facturas[facturas['fecha'] > str(anterior3)]



```

```python
proveedor2 = proveedor[['nombre', 'precio','fecha', 'proveedor']].groupby(['nombre']).agg(fecha = ('fecha','max')).reset_index()
```

```python
proveedor3 = pd.merge(proveedor,proveedor2, on = ['nombre','fecha'], how = 'right')
```

```python

proveedor3
```

```python
fac = facturas_90[["nombre","cantidad","fecha"]].groupby("nombre").agg(fecha_ult_90=("fecha","max"),cantidad_ult_90=("cantidad","sum"))

fac1 = facturas_30[["nombre","cantidad","fecha"]].groupby("nombre").agg(cantidad_ult_30=("cantidad","sum"))

fac2 = facturas_7[["nombre","cantidad","fecha"]].groupby("nombre").agg(cantidad_ult_7=("cantidad","sum"))
```

```python
fac
```

```python
df= pd.merge(items, fac, on = 'nombre', how = 'inner')
```

```python
df2= pd.merge(df, fac1, on = 'nombre', how = 'left')
df3= pd.merge(df2, fac2, on = 'nombre', how = 'left')
```

```python
todo_junto = pd.merge(df3,proveedor3, on = "nombre", how = "left")



```

```python
todo_junto.head()
```

```python
todo_junto1 = todo_junto[['nombre','cantidad_x','proveedor_y','fecha','cantidad','precio','fecha_ult_90','cantidad_ult_90','cantidad_ult_30','cantidad_ult_7']]


```

```python
todo_junto2= todo_junto1.rename(columns={'cantidad_x': 'disponible','cantidad':'ultimo_pedido','proveedor_y':'proveedor','precio':'precio_de_compra','fecha_ult_90':'ultima_venta'})
```

```python
todo_junto2 = todo_junto2.fillna(0)

def para_pedir(row):
    if row['cantidad_ult_30'] > row['disponible']:
        return (row['cantidad_ult_30'] + row['disponible'])
    else:
        return 0
```

```python

todo_junto2['pedir'] = todo_junto2.apply(para_pedir, axis = 1)
```
